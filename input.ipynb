```python
import pandas as pd
df=pd.read_csv("/content/Input.csv",parse_dates=["date"])
```

```python
df
```

Remove duplicates

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

Fill missing attendance status

```python
df.fillna("Absent",inplace=True)
```

```python
df
```

Convert status

```python
df["status"]=df["status"].map({"P":"Present","A":"Absent"})
```

```python
df
```

```python
df.fillna("Absent",inplace=True)
```

```python
df
```

 Generate Summary

```python
total_emp = len(df)
present_emp = (df["status"]=="Present").sum()
absent_emp = (df["status"]=="Absent").sum()
per_attendence = present_emp/total_emp*100
df["date"] = pd.to_datetime(df["date"])
date = df["date"].iloc[0].date()
```

```python
with open("summary.txt","w") as f:
  f.write(f"Date          :{date}\n")
  f.write(f"Total employee: {total_emp}\n")
  f.write(f"Present       : {present_emp}\n")
  f.write(f"Absent        : {absent_emp}\n")
  f.write(f"Attendence %  : {per_attendence}%")
```

```python
with open("summary.txt","r") as f:
  summary = f.read()
print(summary)
```
