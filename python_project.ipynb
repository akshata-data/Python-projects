```
import random
num = random.randint(1,100)
print(num)
count=7
while count>=1:
  x = int(input("guess a number "))
  if x==num:
    print("congrats you won the game")
    a=input("play again?:")
    if a=="Yes":
      print("trial")
      count=7
      num = random.randint(1,100)
      print(num)
      continue
    else:
      print("else")
      break
  else:
    count=count-1
    if count==0:
           print("limit  exceeds")
           break
    else:
      if x>num:
        print("guessed num is too high")
      else:
        print("guessed num is too low")
        continue
```
